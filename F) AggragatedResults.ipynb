# F. Aggregated Results

## 1. Importing Libraries 

```python
# Libraries

import pandas as pd
import numpy as np
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

pd.set_option('display.max_columns', 100)

```

## 2. Results

```python
# Importing Results

LinearModelResults = pd.read_csv('LinearModelsResults.csv', index_col = 'Metric')
BaggingResults = pd.read_csv('BaggingResults.csv', index_col = 'Metric')
RFResults = pd.read_csv('RandomForestResults.csv', index_col = 'Metric')
ABResults = pd.read_csv('AdaBoostResults.csv', index_col = 'Metric')
SVRResults = pd.read_csv('SVRResults.csv', index_col = 'Metric')
```

```python
# Concatenating the Series
Results = pd.concat([LinearModelResults, BaggingResults, RFResults, ABResults, SVRResults], axis = 1 )
```

```python
# Making the dataframe presentable

Results = Results.transpose()
Results.fillna('-------', inplace = True)
Results
```

```python
# Saving the Results
Results.to_csv(r'Results.csv', index=True, index_label = 'Metric')
```

```python
# Case without outliers

# Importing Results

LinearModelsResultsNoOutliers = pd.read_csv('LinearModelsResultsNoOutliers.csv', index_col = 'Metric')
BaggingResultsNoOutliers = pd.read_csv('BaggingResultsNoOutliers.csv', index_col = 'Metric')
RandomForestResultsNoOutliers = pd.read_csv('RandomForestResultsNoOutliers.csv', index_col = 'Metric')
AdaBoostResultsNoOutliers = pd.read_csv('AdaBoostResultsNoOutliers.csv', index_col = 'Metric')
SVRResultsNoOutliers = pd.read_csv('SVRResultsNoOutliers.csv', index_col = 'Metric')
```

```python
# Concatenating the Series
ResultsNoOutliers = pd.concat([LinearModelsResultsNoOutliers, 
                               BaggingResultsNoOutliers, 
                               RandomForestResultsNoOutliers, AdaBoostResultsNoOutliers, 
                               SVRResultsNoOutliers], axis = 1 )
```

```python
# Making the dataframe presentable
ResultsNoOutliers = ResultsNoOutliers.transpose()
ResultsNoOutliers.fillna('-------', inplace = True)
ResultsNoOutliers
```

```python
# Saving the Results
ResultsNoOutliers.to_csv(r'ResultsNoOutliers.csv', index=True, index_label = 'Metric')
```
